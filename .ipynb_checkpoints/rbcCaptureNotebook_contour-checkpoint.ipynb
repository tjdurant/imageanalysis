```python
# csv bound coordinates; min/max of contour. 
import csv 

test = str(rbcImgFiles[1])
image = rbcPath + test

im = cv2.imread(image)

test = im.copy()
test = cv2.cvtColor(test,cv2.COLOR_BGR2GRAY)
test = cv2.medianBlur(test, 5)
ret,thresh = cv2.threshold(test,205,255,0)
image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)

rectParameters = []
for h, cnt in enumerate(contours):
    
    area = cv2.contourArea(cnt)
    if area < 300:
        continue
    if area > 2000:
        continue
    if len(cnt) < 5:
        continue
    
    x,y,w,h = cv2.boundingRect(cnt)   
    param = cv2.boundingRect(cnt) 

    rectParameters.append(param)
    cv2.rectangle(im,(x,y),(x+w,y+h),(0,255,0),2)



    
# cv2.imshow('img',im)
cv2.waitKey(0) 
cv2.imwrite('test.png', im)

myfile = open('rectParams.csv', 'wb')

with myfile:
    csv.register_dialect("custom", delimiter=" ", skipinitialspace=True)
    writer = csv.writer(myfile)
    for tup in rectParameters:
        writer.writerow(tup)
```

```python
###############################################################################
######### Filter
###############################################################################

# http://www.pyimagesearch.com/2014/04/21/building-pokedex-python-finding-game-boy-screen-step-4-6/
# convert image to gray scale
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)

# preprocessing technique for better tresholding?
imgray = cv2.medianBlur(imgray, 5)

edged = cv2.Canny(imgray, 30, 200)
```

```python
image, cnts, hierachy = cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cnts = sorted(cnts, key = cv2.contourArea, reverse = True)[:10]
screenCnt = None
```

```python
# loop over our contours
for c in cnts:
	# approximate the contour
	peri = cv2.arcLength(c, True)
	approx = cv2.approxPolyDP(c, 0.02 * peri, True)
 
	# if our approximated contour has four points, then
	# we can assume that we have found our screen
	if len(approx) == 4:
		screenCnt = approx
		break
```
